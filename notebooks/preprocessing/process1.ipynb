```python
def extract_ar(file):
    sections = file.split('_')
    for section in sections:
        if section.startswith("ar"):
            return section
    return None
```

```python
from pathlib import Path
import pandas as pd
import numpy as np

X = []
y = []
regions = []

for c in classes:
    folder = new_path / split/ c
    
    for csv_samples in folder.glob("*.csv"):
        df = pd.read_csv(csv_samples, sep="\t")

        containsWords = ["FLARE", "LOC", "LABEL", "Timestamp"]

        dropColumns = []

        for col in df.columns:
            for word in containsWords:
                if word in col:
                    dropColumns.append(col)
                    break

        new_df = df.drop(columns=dropColumns)

        X.append(new_df.values)
        y.append(classifyIntensity(csv_samples.name))
        regions.append(extract_ar(csv_samples.name))

X = np.array(X)
y = np.array(y)
regions = np.array(regions)
```
