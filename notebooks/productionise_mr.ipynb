```python
import numpy as np
import pandas as pd
import pickle
```

```python
with open('../models/minirocket.pkl', 'rb') as f:
    minirocket = pickle.load(f)

with open('../models/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open('../models/sgdc.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
def preprocess(sample):
    containsWords = ["FLARE", "LOC", "LABEL", "Timestamp"]

    dropColumns = []

    for col in sample.columns:
        for word in containsWords:
            if word in col:
                dropColumns.append(col)
                break

    sample =  sample.drop(columns=dropColumns)
    return sample.values
```

```python
def prediction(test_sample, threshold = 0.9):
    data = preprocess(test_sample)

    sample = data.reshape(1,60,38)

    X_mr = minirocket.transform(sample)
    X_scaler = scaler.transform(X_mr)

    prob = model.predict_proba(X_scaler)[0,1]
    pred = int(prob >= threshold)

    return pred, prob
```

```python
sample_df = pd.read_csv(
    "../data/raw/partition1/FL/M1.0@265:Primary_ar115_s2010-08-06T08:36:00_e2010-08-06T20:24:00.csv", sep="\t"
)
```

```python
pred, prob = prediction(sample_df)
```
